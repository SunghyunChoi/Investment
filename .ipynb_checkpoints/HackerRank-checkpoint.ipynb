### Interview Preparation Kit > String Manipulation > Special String Again

```python
'''
String S가 주어졌을 때, 해당 String에 특별한 Substring이 몇개나 있는지 찾아라.
특별한 Substring : 모든 Character가 동일하거나 가운데 1개 빼고 동일한 것

1) 각 Character부터 끝까지 돈다.
2) 바로 다음 Character가 동일할 때마다 +1
3) 다른 것이 나올 때 대칭인지 확인
'''

def substrCount(n, s):
    
    answer = 0
    for idx, ch in enumerate(s):
        
        count = 1
        
        for i in range(0, n-idx):
           
            if(ch == s[i + 1]):
                
                count += 1
                continue
            else:
                a = s[i+2:i+2+count]
                b = s[idx:i+1]
                c = s[i+1]
                print(f"s : {s}, a : {a},c : {c}, b : {b}, answer : {answer}")
                
                if(a == b):
                    count += 1
                else : 
                    pass
                
                answer += count
                break
    return answer

s = "mnonopoo"
substrCount(len(s), s)
```

### Interview Preparation Kit > Greedy Algorithms > Luck Balance

```python
# input : k, contests

# Complete the luckBalance function below.
def luckBalance(k, contests):
    
    imp_ct = []
    answer = 0
    
    for ct in contests:
        
        if ct[1] == 1:
            imp_ct.append(ct)
        else:
            answer += ct[0]
            
    imp_ct = sorted(imp_ct, key = lambda x : x[0])
    len_imp = len(imp_ct)
    print(imp_ct)
    count = 0
    win_num = len_imp - k
    
    for ct in imp_ct : 
        
        if(count < win_num):
            answer -= ct[0]
            count += 1
            continue
        else:
            answer += ct[0]
            
    return answer


k = 3
contests = [[5,1], [2,1], [1, 1], [8,1], [10,0], [5,0]]

luckBalance(k, contests)
    
```

```python
from functools import reduce
def transform(num):
    
    if num >= 1000:
        return [num/1000, 1000]
    elif num >= 100:
        return [num/100, 100]
    elif num >= 10:
        return [num/10, 10]
    else:
        return [num, 1]
    
def sums(num):
    return str(int(num[0] * num[1]))
    
def red(x,y):
    return x + y

def solution(numbers):
    answer = ''
    numbers = list(map(transform, numbers))
    print(numbers)
    numbers = sorted(numbers, key= lambda x : x[0], reverse = True)
    print(numbers)
    numbers = list(map(sums, numbers))
    print(numbers)
    answer = reduce(red, numbers)
    
    return answer

solutions([121, 12])
solutions([1000, 0, 0, 0])
solutions([12, 1213])
solutions([21, 212])
solutions([404, 40])


```

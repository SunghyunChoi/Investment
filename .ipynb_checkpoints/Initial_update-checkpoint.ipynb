```python
##Initial Price Update##

#CompanyStockData
#record_date : date 
#company_id int: (index) 
#starting_price: int 
#closing_price: int 
#exchange rate: float 
##Later Update ## market_cap: bigint 


    
#데이터베이스 연결
connection = pymysql.connect(host='localhost',
                             user='investment',
                             password='1111',
                             db='investment',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)

cursor = connection.cursor()
read_company_data = f"SELECT id,code FROM company;"
cursor.execute(read_company_data)

results = cursor.fetchall()
error_sql = []
comp_count = 0
for result in results: 
    
    comp_code, company_id = result['code'], result['id']
    df_daily = fdr.DataReader(comp_code)
    comp_count += 1
    record_dates = []
    
    for idx, row in df_daily.iterrows():
        
        record_date = idx.date()
        record_dates.append(record_date)
        starting_price = row['Open']
        closing_price = row['Close']
        exchange_rate = round(row['Change'],5) if not np.isnan(row['Change']) else 0
        
        #test_sql = f"SELECT code FROM `company` WHERE code = '{code}'"
        
        try:
            sql = f"INSERT INTO `companydailystockdata` (`company_id`, `record_date`, `starting_price`, `closing_price`, `exchange_rate`) VALUES ({company_id}, '{record_date}', {starting_price}, {closing_price}, {exchange_rate});"
            cursor.execute(sql)
        except : 
            error_sql.append(sql)
            continue
            
    print(f"{company_id}, {record_dates[0]} - {record_dates[-1]}, count : {799 - comp_count} left)")
    connection.commit()
    

connection.close()
```

```python
## Initial Update Finance Data ##

'''
+-----------------------+--------------+------+-----+---------+----------------+
| Field                 | Type         | Null | Key | Default | Extra          |
+-----------------------+--------------+------+-----+---------+----------------+
| id                    | int          | NO   | PRI | NULL    | auto_increment |
| record_date           | date         | NO   | UNI | NULL    |                |
| kospi_price           | decimal(8,3) | NO   |     | NULL    |                |
| kospi_exchange_rate    | decimal(7,5) | NO   |     | NULL    |                |
| nasdaq_price          | decimal(9,3) | NO   |     | NULL    |                |
| nasdaq_exchange_rate   | decimal(7,5) | NO   |     | NULL    |                |
| csi300_price          | decimal(9,3) | NO   |     | NULL    |                |
| csi300_exchange_rate   | decimal(7,5) | NO   |     | NULL    |                |
| usd_krw_price         | decimal(9,5) | NO   |     | NULL    |                |
| usd_krw_exchange_rate | decimal(7,5) | NO   |     | NULL    |                |
| gold_price            | decimal(8,3) | NO   |     | NULL    |                |
| gold_exchange_rate     | decimal(7,5) | NO   |     | NULL    |                |
| oil_dbi_price         | decimal(9,6) | NO   |     | NULL    |                |
| oil_dbi_exchange_rate | decimal(7,5) | NO   |     | NULL    |                |
| bitcoin_price         | int          | NO   |     | NULL    |                |
| bitcoin_exchange_rate | decimal(7,5) | NO   |     | NULL    |                |
+-----------------------+--------------+------+-----+---------+----------------+
'''

def daily_finance_data(start_date = None, end_date = None):
    
    #record_date
    timezone = pytz.timezone('Asia/Seoul')
    today = datetime.now(tz = timezone).date()
    
    
    #Check if input format is valid
    if(start_date == None):
        print("Missing start_date : Loading All Data")
    else : 
        try:
            datetime.strptime(start_date, "%Y-%m-%d")
        except : 
            print(f"Not Valid Input Type : start_date {start_date}")
            return 0

    if(end_date == None):
        print(f"Missing end_date : Loading All Data till {today}")
    else : 
        try:
            datetime.strptime(end_date, "%Y-%m-%d")
        except : 
            print(f"Not Valid Input Type : end_date {end_date}")
            return 0

    #KOSPI
    kospi_data = get_data('KS11', start_date, end_date)    
    
    #NASDAQ
    nasdaq_data = get_data('IXIC', start_date, end_date)
    
    #CSI 300
    csi300_data = get_data('CSI300', start_date, end_date)
    
    #USD_KRW
    usd_krw_data = get_data('USD/KRW', start_date, end_date)
    
    #GOLD
    gold_data = get_data('GOLDAMGBD228NLBM', start_date, end_date)
    
    #OIL_DBI
    oil_dbi_data = get_data('OIL_DBI', start_date, end_date)
    
    #BTC/KRW
    btc_data = get_data('BTC/KRW', start_date, end_date)
    
        #데이터베이스 연결
    connection = pymysql.connect(host='localhost',
                                 user='investment',
                                 password='1111',
                                 db='investment',
                                 charset='utf8mb4',
                                 cursorclass=pymysql.cursors.DictCursor)

    connection.close()
    
    df = [kospi_data, nasdaq_data, csi300_data, usd_krw_data, gold_data, oil_dbi_data, btc_data]
    
    return df
    
df_total = daily_finance_data()

#Making Data Range to iterate

date_range_list = []
df_name = ["kospi_data", "nasdaq_data", "csi300_data", "usd_krw_data", "gold_data", "oil_dbi_data", "btc_data"]
min_start = datetime.strptime('2000-01-01', '%Y-%m-%d').date()
max_end = datetime.strptime('2000-01-01', '%Y-%m-%d').date()

for df in df_total:
    
    date_range = df.keys()
    start_date = min(date_range)
    end_date = max(date_range)
    date_range_list.append([start_date, end_date])
    if((start_date - min_start).days < 0):
        min_start = start_date
    if((end_date - max_end).days > 0):
        max_end = end_date                     

for name, date_range in zip(df_name, date_range_list):
    
    print(name, date_range)
    
print(f"start : {min_start}, end : {max_end}")

date_set = []

while min_start <= max_end:
    date_set.append(min_start)
    min_start += relativedelta(days=1)

#Insert data into MySQL

base_sql = f'''INSERT INTO `dailyfinancedata` 
                        (`record_date`, `kospi_price`, `kospi_exchange_rate`, `nasdaq_price`, `nasdaq_exchange_rate`, `csi300_price`, `csi300_exchange_rate`, `usd_krw_price`, `usd_krw_exchange_rate`, `gold_price`, `gold_exchange_rate`, `oil_dbi_price`, `oil_dbi_exchange_rate`, `bitcoin_price`, `bitcoin_exchange_rate`) 
                      VALUES '''


kospi_price, kospi_exchange_rate = 0,0
nasdaq_price, nasdaq_exchange_rate = 0,0
csi300_price,csi300_exchange_rate = 0,0
usd_krw_price, usd_krw_exchange_rate = 0,0
gold_price, gold_exchange_rate = 0,0
oil_dbi_price, oil_dbi_exchange_rate = 0,0
bitcoin_price, bitcoin_exchange_rate = 0,0
count = 0
input_format = []

kospi_data = df_total[0]
nasdaq_data = df_total[1]
csi300_data = df_total[2]
usd_krw_data = df_total[3]
gold_data = df_total[4] 
oil_dbi_data = df_total[5]
bitcoin_data = df_total[6]

def return_data(df, date):
    
    try:
        price = df[date][0]
        exchange_rate = df[date][1]
    except:
        price = None
        exchange_rate = None
    
    return price, exchange_rate

#데이터베이스 연결
connection = pymysql.connect(host='localhost',
                                 user='investment',
                                 password='1111',
                                 db='investment',
                                 charset='utf8mb4',
                                 cursorclass=pymysql.cursors.DictCursor)
test_date_set = date_set

for date in test_date_set:
    
    record_date = datetime.strftime(date, "%Y-%m-%d")
    
    kospi_price, kospi_exchange_rate = return_data(kospi_data, date)
    nasdaq_price, nasdaq_exchange_rate = return_data(nasdaq_data, date)
    csi300_price,csi300_exchange_rate = return_data(csi300_data, date)
    usd_krw_price, usd_krw_exchange_rate = return_data(usd_krw_data, date)
    gold_price, gold_exchange_rate = return_data(gold_data, date)
    oil_dbi_price, oil_dbi_exchange_rate = return_data(oil_dbi_data, date)
    bitcoin_price, bitcoin_exchange_rate = return_data(bitcoin_data, date)
    
    try:
        with connection.cursor() as cursor:
    
            query='''INSERT INTO `dailyfinancedata` (`record_date`, `kospi_price`, `kospi_exchange_rate`, `nasdaq_price`, `nasdaq_exchange_rate`, `csi300_price`, `csi300_exchange_rate`, `usd_krw_price`, `usd_krw_exchange_rate`, `gold_price`, `gold_exchange_rate`, `oil_dbi_price`, `oil_dbi_exchange_rate`, `bitcoin_price`, `bitcoin_exchange_rate`) 
                        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s);
                    '''
            #print(query, (record_date, kospi_price, kospi_exchange_rate, nasdaq_price, nasdaq_exchange_rate, csi300_price, csi300_exchange_rate, usd_krw_price, usd_krw_exchange_rate, gold_price, gold_exchange_rate, oil_dbi_price, oil_dbi_exchange_rate, bitcoin_price, bitcoin_exchange_rate))
            cursor.execute(query, (record_date, kospi_price, kospi_exchange_rate, nasdaq_price, nasdaq_exchange_rate, csi300_price, csi300_exchange_rate, usd_krw_price, usd_krw_exchange_rate, gold_price, gold_exchange_rate, oil_dbi_price, oil_dbi_exchange_rate, bitcoin_price, bitcoin_exchange_rate))

    except : 
        print('error occured', query)
        
    connection.commit()

connection.close()
```

```python
## Volume Update

connection = pymysql.connect(host='localhost',
                             user='investment',
                             password='1111',
                             db='investment',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)
read_company_data = f"SELECT id,code FROM company;"

with connection.cursor() as cursor:
        
    cursor.execute(read_company_data)
    results = cursor.fetchall()
    comp_count = 0

    for result in results: 

        comp_code, company_id = result['code'], result['id']
        df_daily = fdr.DataReader(comp_code)
        comp_count += 1
        day_count = 0
        for idx, row in df_daily.iterrows():
            record_date = idx.date()
            volume = row['Volume']
            #print(f"record_date : {record_date}, comp_code : {comp_code}, data : {row}")
            day_count += 1
            #if(day_count > 10):
            #    sys.exit(0)

            try:
                sql =   '''UPDATE companydailystockdata c
                            SET c.volume = %s
                            WHERE c.company_id = %s and c.record_date = %s;'''


                        #'''SELECT volume 
                        # FROM companydailystockdata c
                        # WHERE c.company_id = %s and c.record_date = %s;'''


                      #'''INSERT INTO `companydailystockdata` (`company_id`, `record_date`, `starting_price`, `closing_price`, `exchange_rate`, `volume`) VALUES (%s, %s, %s, %s, %s, %s)
                      #      ON DUPLICATE KEY UPDATE 
                      #      volume = %s;'''
                cursor.execute(sql, (volume, company_id, record_date))

            except Exception as e: 
                print(e)
        connection.commit()
#print(cursor.fetchall())

connection.close()
```
